```python
from konlpy.tag import Kkma
```

```python
from konlpy.utils import pprint
```

```python
kkma = Kkma()
```

```python
f = open('wiki_ko_sen.txt', 'r')
```

```python
lines = f.readlines()
```

```python
f.close()
```

```python
f = open('morphs_wiki_ko.txt', 'w')
```

```python
input_text = lines[0]
print(input_text)
```

```python
print(len(lines))
```

```python
i = 0
print('start parsing with morphs')
for line in lines:
    
    i+=1
    if i % 10000 == 0:  
        print('Processed ' + str(i) + ' sentences')
    
    input_text = line
    
    output = kkma.pos(input_text)
    for item in output:
        print(item[0], end = ' ', file = f)
    print('',end='\n',file=f)

f.close()
```
